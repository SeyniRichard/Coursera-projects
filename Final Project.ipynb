Is there a significant difference in median value of houses bounded by the Charles river or not?

$𝐻_0: µ_1= µ_2$ ("There is no difference in between the median value of houses bounded by the charles river and median value of houses not bounded by the Charles river")

$𝐻_0: µ_1≠ µ_2$ ("There is a difference in  between the median value of houses bounded by the Charles rivers and the median value of houses not bounded by the Charles river ")

```python
scipy.stats.levene(boston_df[boston_df['CHAS'] == 1]['MEDV'],
                   boston_df[boston_df['CHAS'] == 0]['MEDV'], center='mean')
```

```python
#Since pvalue is lower than 0.05, the variance are not equal
```

```python
#t-test

scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1]['MEDV'],
                  boston_df[boston_df['CHAS'] == 0]['MEDV'], equal_var = False)
```

Conclusion: Since pvalue is less than 0.05, we Reject the null hypothesis thus concluding that there is enough statistical difference between the median value of houses bounded by Charles river and median value of houses not bounded by Charles river.

Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)?

Hypotheses:
    $H_0: µ_1 = µ_2 = µ_3$ (the three age group means are equal)
    
$H_1$ At least one of the means differ

test for equality of variance

```python
thirtyfive_lower_MEDV = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
thirtyfive_seventy_MEDV = boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV']
seventy_older_MEDV = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(thirtyfive_lower_MEDV, thirtyfive_seventy_MEDV, seventy_older_MEDV)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

Conclusion: As Pvalue is greater than 0.05, we fail to reject the null hypothesis as there is no significant evidence that at least one the means differ.

Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town?

Hypotheses:
    $H_0:$ Nitric Acide Concentration is not correlated to the proportion of non-retail business acres per town
    
$H_1:$ Nitric Acide Concentration is correlated to the proportion of non-retail business acres per town

```python
#Pearson Correlation test

scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

Conclusion:  since Pvalue is less than 0.05, we reject the null hypothesis, as there is significant statistical eveidence that there is a correlation between between Nitric Acid Concentration and the proportion of non-retail busiess acres per town

What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes?

```python
import statsmodels.api as sm
```


-   $H_0: β1$ = 0 (Weighted distance to the five Boston employment centres has no effect on the meadian value of owner occupied homes)
-   $H_1: β1$ is not equal to 0 (Weighted distance to the five Boston employment centres has an effect on the meadian value of owner occupied homes)

```python
#Independent Variable
X = boston_df['DIS']
#Dependent Variable
Y = boston_df['MEDV']
#Add Intercept
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

**CONCLUSION:** Since p-value is less than 0.005, we reject the null hypothesis as there is statistical evidence that the Weighted distance to the five Boston employment centres has an effect on the median value of owner occupied homes. And an increase in weighted distance to the five Boston employment centres by one unit increases the median value of owner occupied homes by $1091.6.

**Answer**: **An Additional weighted distance to the five Boston employment centres increases the median value of owner occupied homes by $1091.6**
